```python
# 1st import the package and check its version
import MTM
print("MTM version: ", MTM.__version__)

from MTM import matchTemplates, drawBoxesOnRGB

import cv2
from skimage import io
from skimage.transform import rescale, resize, downscale_local_mean
from skimage.data import coins
import matplotlib.pyplot as plt
```

```python
image = io.imread('photo_2022-08-06_17-07-32.png')
print(image.shape)
plt.imshow(image, cmap="gray")
```

```python
template = io.imread('l-arrow.png')
template = resize(template, (template.shape[0], template.shape[1]),
                       anti_aliasing=True)
plt.imshow(template, cmap="gray")
```

```python
# 1st format the template into a list of tuple (label, templateImage)
listTemplate = [('small', template)]

# Then call the function matchTemplates (here a single template)
Hits = matchTemplates(listTemplate, image, N_object=1, score_threshold=0.3, method=cv2.TM_CCOEFF_NORMED, maxOverlap=0.03)
```

```python
print("Found {} hits".format( len(Hits.index) ) )
Hits
```

```python
Overlay = drawBoxesOnRGB(image, Hits, showLabel=True)
plt.imshow(Overlay)
```
